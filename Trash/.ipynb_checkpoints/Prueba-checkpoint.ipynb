```python
def description(df):
    print(f"Dataset Shape: {df.shape}")
    summary = pd.DataFrame(df.dtypes,columns=['dtypes'])
    summary = summary.reset_index()
    summary['Missing'] = df.isnull().sum().values
    summary['PercMissing'] = df.isnull().sum().values / df.isnull().count().values
    summary['Uniques'] = df.nunique().values
    return summary

description(df)
```

```python
#https://towardsdatascience.com/building-a-comprehensive-set-of-technical-indicators-in-python-for-quantitative-trading-8d98751b5fb
# MovingAverage
values_ma = [5,15,30]

for i in values_ma:
    df["ma_close_" + str(i)] = df["close"].rolling(window=i).mean()
    df["ma_volume_" + str(i)] = df["volume"].rolling(window=i).mean()

# ExponentialMovingWindow
values_ema = [5,15,30]

for i in values_ema:
    df["ema_" + str(i)] = df["close"].ewm(span = i, adjust = False).mean()
    
# RSI https://stackoverflow.com/questions/20526414/relative-strength-index-in-python-pandas
values_rsi = [7,14,21]
close = df['close']

# Get the difference in price from previous step
delta = close.diff()

# Make the positive gains (up) and negative gains (down) Series
up, down = delta.copy(), delta.copy()
up[up < 0] = 0
down[down > 0] = 0

for i in values_rsi:
    # Calculate the EWMA
    roll_up1 = up.ewm(span=i).mean()
    roll_down1 = down.abs().ewm(span=i).mean()

    # Calculate the RSI based on EWMA
    RS1 = roll_up1 / roll_down1
    RSI1 = 100.0 - (100.0 / (1.0 + RS1))

    # Calculate the SMA
    roll_up2 = up.rolling(i).mean()
    roll_down2 = down.abs().rolling(i).mean()

    # Calculate the RSI based on SMA
    RS2 = roll_up2 / roll_down2
    RSI2 = 100.0 - (100.0 / (1.0 + RS2))
    
    df["rsi_ewma_" + str(i)] = RSI1.values
    df["rsi_sma_" + str(i)]  = RSI2.values
    
# BollingerBand
values_bollinger = [5,10,20]
for i in values_bollinger:
    df['bollinger_upper_' + str(i)] = df['close'].rolling(window=i).mean() + (df['close'].rolling(window=i).std()  * 2)
    df['bollinger_lower_' + str(i)] = df['close'].rolling(window=i).mean() - (df['close'].rolling(window=i).std()  * 2)

# Rate of change
values_rc = [5,15,30]
for i in values_rc:
    df["rc_" + str(i)] = df["close"].pct_change(periods = i)
    
# Stochastic Oscillators
values_so = [5,15,30]
for i in values_so:
    lowest = df["low"].rolling(window = i).min()
    high = df["high"].rolling(window = i).max()
    df["so_" + str(i)] = ((df['close'] -lowest)/(high - lowest))*100
    df["so_%D" + str(i)] = df["so_" + str(i)].rolling(window = i).mean()
```

```python
#df["y_high"] = df["high"].shift(-1)
#df["y_low"] = df["low"].shift(-1)
#df[["y_high","high","y_low","low"]].tail()

df["y_close"] = df["close"].shift(-1)
df[["y_close","close"]].tail()
```

```python
past_t = 15 # Es hiperparametro del modelo
minus = 3
```

```python
list_shift = list(range(1, minus))

for i in list_shift: 
    #df["close_minus" + str(i)] = df["close"].shift(i)
    df["high_minus" + str(i)] = df["high"].shift(i)
    df["low_minus" + str(i)] = df["low"].shift(i)
    #df["open_minus" + str(i)] = df["open"].shift(i)
    df["volume_minus" + str(i)] = df["volume"].shift(i)
    df["number_of_trades" + str(i)] = df["number_of_trades"].shift(i)
```

```python
print("Shape: " + str(df.shape))
df.dropna(inplace = True)
print("Shape: " + str(df.shape))
```

```python
X = df.drop(columns=["y_low","y_high"]).copy()
y_low = df["y_low"].copy()
y_high = df["y_high"].copy()
```

```python
from catboost import CatBoostRegressor

modelo = CatBoostRegressor(loss_function='MAE', # Funcion minimizadora del modelo https://catboost.ai/docs/concepts/loss-functions-regression.html
                           verbose = 0,
                           task_type = "CPU" #GPU
                          )

parameters = {'depth' : [2,4,6],
              'learning_rate' : [0.01, 0.1],
              'iterations' : [1000, 2000] }

from sklearn.model_selection import GridSearchCV

grid_catboost = GridSearchCV(estimator = modelo,
                             param_grid = parameters,
                             scoring = "max_error", # https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter
                             #return_train_score = True,
                             cv = 5,
                             n_jobs = -1,
                             verbose = 0              
                            )
```

```python
y_real_low = pd.DataFrame(list(y_low.iloc[past_t:len(y_low)]), columns = ["y_real_low"], index = y_low.iloc[past_t:len(y_low)].index)
y_pred_low = pd.DataFrame([999] * (len(y_low)-past_t), columns = ["y_pred_low"], index = y_low.iloc[past_t:len(y_low)].index)

comparacion_low = pd.concat([y_real_low, y_pred_low], axis =1).copy()

y_real_high = pd.DataFrame(list(y_high.iloc[past_t:len(y_high)]), columns = ["y_real_high"], index = y_high.iloc[past_t:len(y_high)].index)
y_pred_high = pd.DataFrame([999] * (len(y_high)-past_t), columns = ["y_pred_high"], index = y_high.iloc[past_t:len(y_high)].index)

comparacion_high = pd.concat([y_real_high, y_pred_high], axis =1).copy()
```

```python
X.columns
```

```python
tope = len(df)-past_t
df_imp_low = []
df_imp_high = []
for i in range(0, tope):
    # Train
    X_train = X.iloc[i:i+past_t]
    y_train_low = y_low.iloc[i:i+past_t]
    y_train_high = y_high.iloc[i:i+past_t]
    # Test
    X_test = X.iloc[i+past_t]
    y_test_low = y_low.iloc[i+past_t]
    y_test_high = y_high.iloc[i+past_t]
    
    # Train Low
    grid_catboost.fit(X_train, y_train_low)
    pred = grid_catboost.best_estimator_.predict(X_test)
    comparacion_low.at[ y_low.iloc[past_t:len(y_low)].index[i] , 'y_pred_low'] = pred
    fea_imp = pd.DataFrame({'imp': grid_catboost.best_estimator_.feature_importances_, 'col': X_train.columns})
    df_imp_low.append(fea_imp)
    print("Iteration LOW: " + str(i+1) + "/" + str(tope) + " | Date " + str(X_test.name) + " | RealValue " + str(y_test_low) + " | Prediccion " + str(pred), end='\r')
    
    # Train High
    grid_catboost.fit(X_train, y_train_high)
    pred = grid_catboost.best_estimator_.predict(X_test)
    comparacion_high.at[ y_high.iloc[past_t:len(y_high)].index[i] , 'y_pred_high'] = pred
    fea_imp = pd.DataFrame({'imp': grid_catboost.best_estimator_.feature_importances_, 'col': X_train.columns})
    df_imp_high.append(fea_imp)
    print("Iteration HIGH: " + str(i+1) + "/" + str(tope) + " | Date " + str(X_test.name) + " | RealValue " + str(y_test_high) + " | Prediccion " + str(pred), end='\r')   
```

```python
# Variables TOP Low
fea_imp = pd.concat(df_imp_low)
print(fea_imp.shape)
fea_imp = fea_imp.groupby(["col"], as_index =  False).agg({"imp" : 'mean'})
print(fea_imp.shape)
fea_imp = fea_imp.sort_values(['imp', 'col'], ascending=[True, False])#.iloc[-20:]
_ = fea_imp.plot(kind='barh', x='col', y='imp', figsize=(20, 20))
```

```python
# Variables TOP High
fea_imp = pd.concat(df_imp_high)
print(fea_imp.shape)
fea_imp = fea_imp.groupby(["col"], as_index =  False).agg({"imp" : 'mean'})
print(fea_imp.shape)
fea_imp = fea_imp.sort_values(['imp', 'col'], ascending=[True, False])#.iloc[-20:]
_ = fea_imp.plot(kind='barh', x='col', y='imp', figsize=(20, 20))
```
